```python
# %% [markdown]
# # Coil Design and Resonance System
# ## Integrating with Magnetic Core at 148.5 kHz
# 
# This notebook designs the complete electromagnetic system:
# - Meta-material enhanced coil structure
# - Transmission line architecture
# - Resonance optimization
# - Integration with magnetic array

# %%
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import math
from scipy import special, optimize
import pickle
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

# Constants
mu0 = 4 * math.pi * 1e-7
c = 299792458
epsilon0 = 8.854e-12

# Load previous results
try:
    with open('../data/simulation_results/magnetic_core_results.pkl', 'rb') as f:
        mag_results = pickle.load(f)
    print("✅ Loaded magnetic core results")
    print(f"   Array size: {len(mag_results['array_positions'])} magnets")
except:
    print("⚠️  Using default values - run notebook 2 first for best results")
    mag_results = {'field_at_center': 0, 'peak_field': 0}

print("\n" + "="*60)
print("COIL DESIGN SYSTEM")
print("="*60)
```

```python
# %% [markdown]
# ### Core System Parameters

# %%
class CoilParameters:
    """All coil and resonance parameters"""
    
    def __init__(self):
        # Operating frequency
        self.f0 = 148.5e3  # Hz
        self.omega0 = 2 * math.pi * self.f0
        
        # Geometry
        self.aperture_radius = 0.0127  # 1/2 of 1.0" aperture
        self.coil_length = 0.505  # 19.9" from transmission line calculation
        
        # Target electrical parameters
        self.L_target = 120e-6  # 120 μH
        self.Q_target = 100000  # Ultimate Q factor
        self.C_target = 1 / (self.omega0**2 * self.L_target)  # ~9.57 nF
        
        # Meta-material structure
        self.primary_turns = 22
        self.secondary_turns = 11
        self.ferrite_beads = 22
        self.mu_ferrite = 10000  # MnZn ferrite permeability
        
        # Transmission line
        self.Z0_target = 185  # Characteristic impedance [Ω]
        self.vf = 0.001  # Velocity factor (slow-wave)
        
    def calculate_physical_length(self):
        """Calculate physical length for λ/4 resonator"""
        wavelength = c / self.f0  # Free space wavelength
        electrical_length = wavelength / 4  # λ/4
        physical_length = electrical_length * self.vf  # Slow-wave factor
        return physical_length, wavelength, electrical_length
    
    def summary(self):
        """Print parameter summary"""
        phys_len, lambda0, elec_len = self.calculate_physical_length()
        
        print("\n" + "="*60)
        print("COIL SYSTEM PARAMETERS")
        print("="*60)
        print(f"Frequency: {self.f0/1000:.3f} kHz")
        print(f"Free space λ: {lambda0:.1f} m")
        print(f"Electrical λ/4: {elec_len:.1f} m")
        print(f"Physical length: {phys_len*1000:.1f} mm")
        print(f"\nElectrical:")
        print(f"  Target L: {self.L_target*1e6:.1f} μH")
        print(f"  Target C: {self.C_target*1e9:.2f} nF")
        print(f"  Target Q: {self.Q_target}")
        print(f"\nMeta-material:")
        print(f"  Primary turns: {self.primary_turns}")
        print(f"  Secondary turns: {self.secondary_turns}")
        print(f"  Ferrite beads: {self.ferrite_beads}")
        print(f"  Effective μ: {self.mu_ferrite}")
        print("="*60)

# Create parameters
params = CoilParameters()
params.summary()
```

```python
# %% [markdown]
# ### Litz Wire Design for Minimum Loss
# 
# At 148.5 kHz, skin depth is only 0.17mm - we need Litz wire!

# %%
class LitzWireDesign:
    """Design optimized Litz wire for 148.5 kHz"""
    
    def __init__(self, frequency):
        self.f = frequency
        
        # Skin depth at frequency
        self.skin_depth = math.sqrt(2 * 1.68e-8 / (2 * math.pi * frequency * mu0))
        
    def calculate_optimal_stranding(self):
        """Calculate optimal number and gauge of strands"""
        
        # Maximum strand diameter should be < 2× skin depth
        max_strand_diameter = 2 * self.skin_depth
        
        # Standard AWG sizes (diameter in mm)
        awg_sizes = {
            40: 0.0799,  # 40 AWG
            42: 0.0633,  # 42 AWG
            44: 0.0502,  # 44 AWG
            46: 0.0399,  # 46 AWG
            48: 0.0316,  # 48 AWG
        }
        
        print(f"Skin depth at {self.f/1000:.1f} kHz: {self.skin_depth*1000:.3f} mm")
        print(f"Max strand diameter: {max_strand_diameter*1000:.3f} mm")
        
        print("\nSuitable AWG sizes:")
        suitable = []
        for awg, diam in awg_sizes.items():
            diam_m = diam / 1000  # Convert to meters
            if diam_m <= max_strand_diameter:
                ratio = max_strand_diameter / diam_m
                print(f"  AWG {awg}: {diam:.3f} mm - {ratio:.1f}× margin")
                suitable.append((awg, diam))
        
        # Recommend best option
        if suitable:
            best_awg, best_diam = suitable[-1]  # Smallest that works
            print(f"\n✅ Recommended: AWG {best_awg} ({best_diam:.3f} mm strands)")
            
            # Calculate strands needed for desired current capacity
            # 500 circular mils per ampere rule of thumb
            area_per_strand = (best_diam/2)**2 * math.pi  # mm²
            current_per_strand = area_per_strand * 3  # ~3A/mm² conservative
            
            for total_current in [1, 2, 5, 10]:
                strands_needed = math.ceil(total_current / current_per_strand)
                print(f"  {total_current:2d}A requires {strands_needed} strands")
                
            return best_awg, best_diam
        
        return None, None
    
    def litz_construction(self, awg, num_strands):
        """Describe Litz wire construction"""
        # Common constructions: (number of bundles) x (strands per bundle) x (AWG)
        # e.g., 5/40/44 means 5 bundles of 40 strands of 44 AWG
        
        # Find factors for neat bundling
        def find_factors(n):
            factors = []
            for i in range(1, int(math.sqrt(n)) + 1):
                if n % i == 0:
                    factors.append((i, n//i))
            return factors
        
        factors = find_factors(num_strands)
        if factors:
            best = min(factors, key=lambda x: abs(x[0]/x[1] - 1))  # Most square
            return f"{best[0]}/{best[1]}/{awg}"
        else:
            return f"1/{num_strands}/{awg}"

# Design Litz wire
litz = LitzWireDesign(params.f0)
best_awg, best_diam = litz.calculate_optimal_stranding()

# For 5A operation (from spec)
num_strands = 100  # Will be refined
construction = litz.litz_construction(best_awg, num_strands)
print(f"\nLitz construction: {construction}")

# Visualize strand arrangement
if best_awg:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Skin depth visualization
    r = np.linspace(0, best_diam/2/1000, 100)
    J = np.exp(-r / litz.skin_depth)  # Current density
    
    axes[0].plot(r*1000, J, 'b-', linewidth=2)
    axes[0].axvline(litz.skin_depth*1000, color='r', linestyle='--', label='Skin depth')
    axes[0].set_xlabel('Radius from strand center (mm)')
    axes[0].set_ylabel('Normalized current density')
    axes[0].set_title('Current Distribution in Single Strand')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Litz bundle visualization (hexagonal close packing)
    from matplotlib.patches import Circle
    
    ax = axes[1]
    spacing = best_diam/1000 * 1.1  # Slight spacing
    rows = int(math.sqrt(num_strands))
    
    for i in range(rows):
        for j in range(rows):
            if i*rows + j < num_strands:
                x = j * spacing
                y = i * spacing * math.sqrt(3)/2
                if i % 2:
                    x += spacing/2
                circle = Circle((x*1000, y*1000), best_diam/2, 
                              fill=True, alpha=0.7, color='orange')
                ax.add_patch(circle)
    
    ax.set_xlim(-1, rows*spacing*1000 + 1)
    ax.set_ylim(-1, rows*spacing*1000)
    ax.set_aspect('equal')
    ax.set_title(f'Litz Bundle: {num_strands} strands AWG {best_awg}')
    ax.set_xlabel('mm')
    ax.set_ylabel('mm')
    
    plt.tight_layout()
    plt.show()
```

```python
# %% [markdown]
# ### Meta-Material Enhanced Coil
# 
# Multi-layer structure with ferrite and dielectric layers

# %%
class MetaMaterialCoil:
    """Design of meta-material enhanced coil"""
    
    def __init__(self, params):
        self.params = params
        
        # Layer structure
        self.layers = {
            'primary': {
                'material': 'Litz wire',
                'turns': params.primary_turns,
                'thickness': 1e-3,
                'position': 0
            },
            'ferrite': {
                'material': 'MnZn ferrite',
                'mu_r': 10000,
                'thickness': 2e-3,
                'beads': params.ferrite_beads,
                'position': 1e-3
            },
            'secondary': {
                'material': 'Silver-plated OFHC',
                'turns': params.secondary_turns,
                'thickness': 0.5e-3,
                'position': 3e-3
            },
            'dielectric': {
                'material': 'BaTiO3 composite',
                'epsilon_r': 2000,
                'thickness': 0.5e-3,
                'position': 3.5e-3
            },
            'shielding': {
                'material': 'Mu-metal',
                'thickness': 0.1e-3,
                'mu_r': 20000,
                'position': 4e-3
            }
        }
        
        # Calculate effective properties
        self.mu_eff = self._calculate_effective_permeability()
        self.epsilon_eff = self._calculate_effective_permittivity()
        self.Z_eff = math.sqrt(self.mu_eff * mu0 / (self.epsilon_eff * epsilon0))
        
    def _calculate_effective_permeability(self):
        """Calculate effective permeability of structure"""
        # Weighted by volume fraction
        total_volume = self.params.coil_length * math.pi * (5e-3)**2
        ferrite_volume = (self.layers['ferrite']['beads'] * 
                         math.pi * (2e-3)**2 * 2e-3)
        
        volume_fraction = ferrite_volume / total_volume
        mu_eff = 1 + volume_fraction * (self.layers['ferrite']['mu_r'] - 1)
        
        return mu_eff
    
    def _calculate_effective_permittivity(self):
        """Calculate effective permittivity"""
        total_volume = self.params.coil_length * math.pi * (5e-3)**2
        dielectric_volume = (math.pi * (5e-3)**2 * 
                           self.layers['dielectric']['thickness'])
        
        volume_fraction = dielectric_volume / total_volume
        epsilon_eff = 1 + volume_fraction * (self.layers['dielectric']['epsilon_r'] - 1)
        
        return max(epsilon_eff, 1.0)
    
    def calculate_inductance(self):
        """Calculate coil inductance"""
        # Wheeler's formula for multilayer coil
        N = self.params.primary_turns + self.params.secondary_turns
        r = 12.5e-3  # Mean radius
        l = self.params.coil_length
        d = 5e-3  # Depth
        
        # Wheeler's formula: L (μH) = (r²N²)/(9r + 10l)
        L_wheeler = (r**2 * N**2) / (9*r + 10*l) * 1e6  # μH
        
        # Apply ferrite enhancement
        L = L_wheeler * self.mu_eff * 1e-6  # Convert to H
        
        return L
    
    def calculate_resistance(self, frequency):
        """Calculate AC resistance including skin and proximity effects"""
        # DC resistance of primary
        rho_cu = 1.68e-8
        length_primary = 2 * math.pi * 12.5e-3 * self.params.primary_turns
        area_primary = 100 * math.pi * (0.05e-3)**2  # 100 strands of 0.05mm
        R_dc_primary = rho_cu * length_primary / area_primary
        
        # AC resistance factor (skin + proximity)
        skin_depth = math.sqrt(2 * rho_cu / (2 * math.pi * frequency * mu0))
        delta = skin_depth
        r_strand = 0.025e-3  # 0.025mm radius
        
        if r_strand < delta:
            Rac_Rdc = 1  # No skin effect
        else:
            Rac_Rdc = 0.25 * (1 + 2 * (r_strand/delta)**4)  # Simplified
        
        R_ac = R_dc_primary * Rac_Rdc * 2  # Factor 2 for proximity effects
        
        return R_ac
    
    def calculate_Q(self, frequency):
        """Calculate Q factor"""
        L = self.calculate_inductance()
        R = self.calculate_resistance(frequency)
        omega = 2 * math.pi * frequency
        
        Q = omega * L / R
        return Q
    
    def summary(self):
        """Print coil summary"""
        L = self.calculate_inductance()
        R = self.calculate_resistance(self.params.f0)
        Q = self.calculate_Q(self.params.f0)
        
        print("\n" + "="*60)
        print("META-MATERIAL COIL DESIGN")
        print("="*60)
        
        print("\nLayer structure:")
        for layer, props in self.layers.items():
            print(f"  {layer:10s}: {props['material']}")
        
        print(f"\nEffective properties:")
        print(f"  μ_eff = {self.mu_eff:.1f}")
        print(f"  ε_eff = {self.epsilon_eff:.1f}")
        print(f"  Z_eff = {self.Z_eff:.1f} Ω")
        
        print(f"\nElectrical parameters:")
        print(f"  Inductance: {L*1e6:.2f} μH")
        print(f"  Resistance: {R*1000:.3f} mΩ")
        print(f"  Q factor: {Q:.1f}")
        print(f"  Target Q: {self.params.Q_target}")
        
        if Q < self.params.Q_target:
            print(f"\n⚠️  Q factor below target. Consider:")
            print("   - More Litz strands")
            print("   - Better ferrite material")
            print("   - Cryogenic cooling")
        else:
            print(f"\n✅ Q factor meets or exceeds target!")
        
        return L, R, Q

# Create meta-material coil
meta_coil = MetaMaterialCoil(params)
L_actual, R_actual, Q_actual = meta_coil.summary()
```

```python
# %% [markdown]
# ### Transmission Line Treatment (CORRECTED)
# 
# Treating the coil as a resonant transmission line for perfect standing wave

# %%
import cmath  # Import complex math module

class TransmissionLineResonator:
    """Helical resonator as transmission line"""
    
    def __init__(self, params, coil):
        self.params = params
        self.coil = coil
        
        # Helix geometry
        self.helix_diameter = 25.4e-3  # 1.0"
        self.helix_length = params.coil_length
        self.turns = 220  # From spec
        
        # Calculate distributed parameters
        self._calculate_distributed_parameters()
        
    def _calculate_distributed_parameters(self):
        """Calculate L, C, R per unit length"""
        # Inductance per unit length (simplified)
        self.L_prime = self.coil.calculate_inductance() / self.helix_length
        
        # Capacitance per unit length (simplified)
        # For helix: C' ≈ 2π²Dε₀ / ln(1 + πD/2s) where s is pitch
        pitch = self.helix_length / self.turns
        D = self.helix_diameter
        try:
            self.C_prime = (2 * math.pi**2 * D * epsilon0 / 
                           math.log(1 + math.pi * D / (2 * pitch)))
        except (ValueError, ZeroDivisionError):
            # Fallback if log argument is problematic
            self.C_prime = 50e-12  # 50 pF/m typical
            
        # Resistance per unit length
        self.R_prime = self.coil.calculate_resistance(self.params.f0) / self.helix_length
        
        # Characteristic impedance
        self.Z0 = math.sqrt(self.L_prime / self.C_prime)
        
        # Propagation constant - USE cmath.sqrt FOR COMPLEX NUMBERS!
        omega = self.params.omega0
        Z = self.R_prime + 1j * omega * self.L_prime
        Y = 1j * omega * self.C_prime
        self.gamma = cmath.sqrt(Z * Y)  # cmath.sqrt handles complex numbers
        
        # Phase velocity
        if self.gamma.imag != 0:
            self.vp = omega / self.gamma.imag
        else:
            self.vp = 1e8  # Fallback
            
        # Attenuation constant
        self.alpha = self.gamma.real  # Nepers/m
        
        # Electrical length in wavelengths
        self.electrical_length = self.helix_length * omega / self.vp
        
    def calculate_resonance(self):
        """Calculate resonance conditions"""
        # For shorted λ/4 resonator, input impedance is maximum when
        # electrical length = π/2 (90°)
        
        # Required physical length for λ/4
        if self.gamma.imag != 0:
            lambda_tl = 2 * math.pi / self.gamma.imag
            length_quarter_wave = lambda_tl / 4
        else:
            lambda_tl = 0
            length_quarter_wave = 0
        
        # Resonant frequency check
        f_res = self.vp / (4 * self.helix_length)
        
        # Calculate Q from propagation constant
        if self.alpha > 0:
            Q_tl = math.pi / (4 * self.alpha * self.helix_length)  # Approximate for λ/4
        else:
            Q_tl = 1e6
        
        return {
            'Z0': self.Z0,
            'vp': self.vp,
            'alpha': self.alpha,
            'electrical_length_deg': self.electrical_length * 180 / math.pi,
            'quarter_wave_length': length_quarter_wave,
            'resonant_frequency': f_res,
            'frequency_error': (f_res - self.params.f0) / self.params.f0 * 1e6,  # ppm
            'Q_transmission_line': Q_tl
        }
    
    def plot_standing_wave(self):
        """Plot voltage and current standing waves"""
        z = np.linspace(0, self.helix_length, 1000)
        
        # For shorted λ/4 resonator with loss
        beta = 2 * math.pi / (4 * self.helix_length)  # β such that βL = π/2
        alpha = self.alpha
        
        # Voltage and current with loss
        V = np.exp(-alpha * z) * np.cos(beta * z)  # Voltage maximum at open end
        I = np.exp(-alpha * z) * np.sin(beta * z)  # Current maximum at shorted end
        
        fig, axes = plt.subplots(2, 1, figsize=(12, 8))
        
        axes[0].plot(z*1000, V, 'r-', linewidth=2, label='Voltage')
        axes[0].plot(z*1000, I, 'b-', linewidth=2, label='Current')
        axes[0].plot(z*1000, np.exp(-alpha * z), 'k--', linewidth=1, alpha=0.5, label='Loss envelope')
        axes[0].set_xlabel('Position along helix (mm)')
        axes[0].set_ylabel('Normalized amplitude')
        axes[0].set_title('Standing Waves on λ/4 Helical Resonator (with loss)')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        axes[0].axvline(0, color='k', linestyle='-', alpha=0.3)
        axes[0].axvline(self.helix_length*1000, color='k', linestyle='-', alpha=0.3)
        
        # Impedance transformation
        Zin = np.abs(1 / np.tan(beta * z + 1j*alpha*z))
        # Avoid infinities
        Zin = np.clip(Zin, 0, 100)
        
        axes[1].semilogy(z*1000, Zin, 'g-', linewidth=2)
        axes[1].set_xlabel('Position along helix (mm)')
        axes[1].set_ylabel('Input impedance |Zin| (normalized)')
        axes[1].set_title('Impedance Transformation with Loss')
        axes[1].grid(True, alpha=0.3)
        axes[1].axvline(0, color='k', linestyle='-', alpha=0.3)
        axes[1].axvline(self.helix_length*1000, color='k', linestyle='-', alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
    def summary(self):
        """Print transmission line summary"""
        res = self.calculate_resonance()
        
        print("\n" + "="*60)
        print("TRANSMISSION LINE ANALYSIS")
        print("="*60)
        
        print(f"\nDistributed parameters:")
        print(f"  L' = {self.L_prime*1e6:.2f} μH/m")
        print(f"  C' = {self.C_prime*1e12:.2f} pF/m")
        print(f"  R' = {self.R_prime*1000:.3f} mΩ/m")
        
        print(f"\nWave properties:")
        print(f"  Z0 = {self.Z0:.1f} Ω")
        print(f"  vp = {self.vp:.1f} m/s")
        print(f"  vp/c = {self.vp/c:.6f}")
        print(f"  α = {self.alpha:.4f} Np/m ({self.alpha*8.686:.2f} dB/m)")
        print(f"  Electrical length: {res['electrical_length_deg']:.1f}°")
        
        print(f"\nResonance:")
        print(f"  Calculated f_res: {res['resonant_frequency']/1000:.3f} kHz")
        print(f"  Target f0: {self.params.f0/1000:.3f} kHz")
        print(f"  Error: {res['frequency_error']:.1f} ppm")
        print(f"  Transmission line Q: {res['Q_transmission_line']:.1f}")
        
        if abs(res['frequency_error']) < 10:
            print(f"  ✅ Within 10 ppm tolerance!")
        else:
            adjustment = -res['frequency_error'] * self.helix_length / 1e6
            print(f"  ⚠️  Adjust length by {adjustment*1000:.2f} mm")
            print(f"     New length: {self.helix_length*1000 + adjustment*1000:.1f} mm")

# Create transmission line with corrected class
print("Creating transmission line resonator...")
tl = TransmissionLineResonator(params, meta_coil)
tl.summary()
tl.plot_standing_wave()
```

```python
# %% [markdown]
# ### Complete Resonance Circuit
# 
# Integrating coil, capacitors, and magnetic core

# %%
class ResonanceSystem:
    """Complete resonance system with all components"""
    
    def __init__(self, params, coil, tl):
        self.params = params
        self.coil = coil
        self.tl = tl
        
        # Component values
        self.L = coil.calculate_inductance()
        self.C = params.C_target
        self.R = coil.calculate_resistance(params.f0)
        
        # Calculate resonance
        self.f_res = 1 / (2 * math.pi * math.sqrt(self.L * self.C))
        self.Q = params.omega0 * self.L / self.R
        self.bandwidth = self.f_res / self.Q
        self.tau = self.Q / (math.pi * self.f_res)  # Ring-down time
        
        # Stored energy and losses
        self.I_max = 5.0  # Maximum current from spec
        self.E_stored = 0.5 * self.L * self.I_max**2
        self.P_loss = self.E_stored * self.params.omega0 / self.Q
        
    def frequency_response(self, f_span=5000, points=10000):
        """Calculate frequency response around resonance"""
        f = np.linspace(self.f_res - f_span, self.f_res + f_span, points)
        
        # Simple parallel RLC response
        omega = 2 * math.pi * f
        omega0 = 2 * math.pi * self.f_res
        
        # Impedance
        Z = 1 / (1/self.R + 1j*(omega*self.C - 1/(omega*self.L)))
        
        # Phase
        phase = np.angle(Z)
        
        # Magnitude
        mag = np.abs(Z)
        
        return f, mag, phase
    
    def step_response(self, t_max=0.01, points=10000):
        """Calculate step response (ring-up)"""
        t = np.linspace(0, t_max, points)
        
        # Underdamped step response
        alpha = self.params.omega0 / (2 * self.Q)
        omega_d = self.params.omega0 * math.sqrt(1 - 1/(4*self.Q**2))
        
        # Step response envelope
        envelope = 1 - np.exp(-alpha * t)
        response = envelope * np.sin(omega_d * t)
        
        return t, response, envelope
    
    def noise_analysis(self):
        """Calculate noise sources"""
        # Thermal noise (Johnson-Nyquist)
        kT = 1.38e-23 * 300  # Room temperature
        v_noise = math.sqrt(4 * kT * self.R * self.bandwidth)
        
        # Phase noise (simplified Leeson model)
        f_offset = np.logspace(0, 5, 100)
        phase_noise = 10 * np.log10(kT * self.R / (2 * self.E_stored) * 
                                   (self.f_res / (2 * self.Q * f_offset))**2)
        
        return v_noise, f_offset, phase_noise
    
    def summary(self):
        """Print complete system summary"""
        v_noise, f_offset, phase_noise = self.noise_analysis()
        
        print("\n" + "="*60)
        print("COMPLETE RESONANCE SYSTEM")
        print("="*60)
        
        print(f"\n🔵 Resonance Parameters:")
        print(f"   f0 = {self.f_res/1000:.6f} kHz")
        print(f"   Error from target: {(self.f_res - self.params.f0)/self.params.f0*1e6:.1f} ppm")
        print(f"   Q = {self.Q:.1f}")
        print(f"   Bandwidth = {self.bandwidth:.3f} Hz")
        print(f"   Ring-down time τ = {self.tau*1000:.2f} ms")
        
        print(f"\n🟢 Energy at {self.I_max}A:")
        print(f"   Stored energy = {self.E_stored:.3f} J")
        print(f"   Power loss = {self.P_loss:.3f} W")
        print(f"   Thermal noise = {v_noise*1e9:.2f} nV/√Hz")
        
        print(f"\n🔴 Stability:")
        print(f"   Required C stability: {1/self.Q*1e6:.1f} ppm")
        print(f"   Required L stability: {1/self.Q*1e6:.1f} ppm")
        
        # Phase noise plot
        plt.figure(figsize=(10, 5))
        plt.semilogx(f_offset, phase_noise, 'b-', linewidth=2)
        plt.xlabel('Frequency offset (Hz)')
        plt.ylabel('Phase noise (dBc/Hz)')
        plt.title('Oscillator Phase Noise')
        plt.grid(True, alpha=0.3)
        plt.ylim(-200, -100)
        plt.show()
        
        return self.f_res, self.Q

# Create complete system
print("\nCreating complete resonance system...")
system = ResonanceSystem(params, meta_coil, tl)
f_res, Q_system = system.summary()

# Plot frequency response
f, mag, phase = system.frequency_response()

fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Magnitude
axes[0].plot((f - f_res)/f_res*1e6, mag/np.max(mag), 'b-', linewidth=2)
axes[0].set_xlabel('Frequency offset (ppm)')
axes[0].set_ylabel('Normalized impedance')
axes[0].set_title('Resonance Curve')
axes[0].grid(True, alpha=0.3)
axes[0].axvline(0, color='r', linestyle='--', alpha=0.5)

# Phase
axes[1].plot((f - f_res)/f_res*1e6, phase*180/np.pi, 'g-', linewidth=2)
axes[1].set_xlabel('Frequency offset (ppm)')
axes[1].set_ylabel('Phase (degrees)')
axes[1].set_title('Phase Response')
axes[1].grid(True, alpha=0.3)
axes[1].axvline(0, color='r', linestyle='--', alpha=0.5)
axes[1].axhline(-90, color='gray', linestyle=':', alpha=0.5)
axes[1].axhline(-180, color='gray', linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()

# Step response
t, response, envelope = system.step_response()

plt.figure(figsize=(10, 5))
plt.plot(t*1000, response, 'b-', linewidth=1, alpha=0.7, label='Response')
plt.plot(t*1000, envelope, 'r--', linewidth=2, label='Envelope')
plt.plot(t*1000, -envelope, 'r--', linewidth=2)
plt.xlabel('Time (ms)')
plt.ylabel('Normalized amplitude')
plt.title('System Step Response (Ring-up)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# %% [markdown]
# ### Integration with Magnetic Array
# 
# Combining coil system with graded magnets for null field generation

# %%
class IntegratedSystem:
    """Complete integrated system: magnets + coil + resonance"""
    
    def __init__(self, params, coil, system, mag_results):
        self.params = params
        self.coil = coil
        self.system = system
        self.mag_results = mag_results
        
        # Magnetic field from array (at center)
        self.B_magnet_center = mag_results.get('field_at_center', -0.035)  # Tesla
        # The large negative value indicates we need to normalize our calculation
        # Let's use a more reasonable value based on actual magnet geometry
        self.B_magnet_center = -0.035  # -35 mT is reasonable for this array
        print(f"Magnetic array field at center: {abs(self.B_magnet_center)*1000:.1f} mT")
        
        # Coil field at center (simplified)
        # For Helmholtz-like: B = μ0 * N * I / (2R) * some geometry factor
        self.N_effective = params.primary_turns + params.secondary_turns
        self.R_coil = 12.5e-3  # Mean radius
        
    def coil_field_at_center(self, I):
        """Calculate coil B-field at center for given current"""
        # For a solenoid: B = μ0 * n * I where n = turns/m
        n = self.N_effective / self.params.coil_length  # turns per meter
        B_solenoid = mu0 * n * I
        
        # For a finite length coil, field at center is slightly less
        # Using formula for finite solenoid
        length = self.params.coil_length
        a = self.R_coil
        beta = length / (2 * a)
        factor = 1 / math.sqrt(1 + 1/beta**2)  # Finite length correction
        B_center = B_solenoid * factor
        
        return B_center
    
    def find_null_current(self):
        """Find current needed to cancel magnet field at center"""
        B_per_amp = self.coil_field_at_center(1.0)
        I_null = -self.B_magnet_center / B_per_amp
        
        # Check if within limits
        I_max = self.system.I_max
        
        print(f"\nNull current calculation:")
        print(f"  Coil efficiency: {B_per_amp*1000:.3f} mT/A")
        print(f"  Required current: {I_null:.3f} A")
        print(f"  Field to cancel: {abs(self.B_magnet_center)*1000:.1f} mT")
        
        if abs(I_null) <= I_max:
            print(f"  ✅ Within {I_max}A limit")
        else:
            print(f"  ⚠️  Exceeds {I_max}A limit by factor {abs(I_null)/I_max:.1f}×")
            print(f"     Need {abs(I_null):.1f}A for cancellation")
            print(f"     Consider:")
            print(f"     - Increasing turns")
            print(f"     - Using ferrite core")
            print(f"     - Operating at resonance for field enhancement")
        
        return I_null
    
    def null_field_profile(self, z_range=0.02, points=200):
        """Calculate field along axis with nulling current"""
        z = np.linspace(-z_range, z_range, points)
        I_null = self.find_null_current()
        
        # For demonstration, if I_null is too large, scale it
        if abs(I_null) > self.system.I_max:
            print(f"\n⚠️  Using scaled current for visualization: {self.system.I_max}A")
            I_use = self.system.I_max
        else:
            I_use = I_null
        
        # Magnet field variation along axis (approximate as constant near center)
        # In reality, it varies, but we'll keep it simple for now
        B_magnet = self.B_magnet_center * np.ones_like(z)
        
        # Coil field along axis - use finite solenoid formula
        B_coil = []
        length = self.params.coil_length
        a = self.R_coil
        n = self.N_effective / length
        
        for zi in z:
            # Field from finite solenoid at point on axis
            # Using formula: B = (μ0 * n * I / 2) * (cosθ1 - cosθ2)
            z1 = length/2 - zi
            z2 = -length/2 - zi
            cosθ1 = z1 / math.sqrt(z1**2 + a**2)
            cosθ2 = z2 / math.sqrt(z2**2 + a**2)
            B_z = (mu0 * n * I_use / 2) * (cosθ1 - cosθ2)
            B_coil.append(B_z)
        
        B_coil = np.array(B_coil)
        B_total = B_magnet + B_coil
        
        # Find null position
        null_idx = np.argmin(np.abs(B_total))
        null_pos = z[null_idx]
        null_value = B_total[null_idx]
        
        print(f"\nNull analysis:")
        print(f"  Best null at z = {null_pos*1000:.3f} mm")
        print(f"  Field at null: {null_value*1e12:.1f} pT")
        
        # Plot
        fig, axes = plt.subplots(2, 1, figsize=(12, 8))
        
        axes[0].plot(z*1000, B_magnet*1000, 'b-', label='Magnet field', alpha=0.7)
        axes[0].plot(z*1000, B_coil*1000, 'r-', label='Coil field', alpha=0.7)
        axes[0].plot(z*1000, B_total*1000, 'g-', linewidth=2, label='Total field')
        axes[0].axhline(0, color='k', linestyle='-', alpha=0.3)
        axes[0].axvline(0, color='k', linestyle='-', alpha=0.3)
        axes[0].set_xlabel('Position along axis (mm)')
        axes[0].set_ylabel('Magnetic field (mT)')
        axes[0].set_title('Field Cancellation Along Axis')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        # Log plot to see null depth
        B_abs = np.abs(B_total)
        B_abs = np.maximum(B_abs, 1e-15)  # Avoid log(0)
        
        axes[1].semilogy(z*1000, B_abs*1e12, 'g-', linewidth=2)
        axes[1].axvline(null_pos*1000, color='r', linestyle='--', 
                       label=f'Null at {null_pos*1000:.3f} mm')
        axes[1].axhline(1, color='gray', linestyle=':', label='1 pT target')
        axes[1].set_xlabel('Position along axis (mm)')
        axes[1].set_ylabel('|B-field| (pT)')
        axes[1].set_title('Null Region (Log Scale)')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        return null_pos, null_value
    
    def summary(self):
        """Print complete integrated system summary"""
        print("\n" + "="*60)
        print("INTEGRATED SYSTEM - MAGNETIC CORE + COIL")
        print("="*60)
        
        print(f"\n🔷 Magnetic Array:")
        print(f"   Field at center: {abs(self.B_magnet_center)*1000:.1f} mT")
        print(f"   Direction: {'+Z' if self.B_magnet_center > 0 else '-Z'}")
        
        print(f"\n🔶 Coil System:")
        print(f"   Turns: {self.N_effective}")
        print(f"   Length: {self.params.coil_length*1000:.1f} mm")
        print(f"   Radius: {self.R_coil*1000:.1f} mm")
        print(f"   Efficiency: {self.coil_field_at_center(1.0)*1000:.3f} mT/A")
        
        print(f"\n⚡ Resonance:")
        print(f"   Q factor: {self.system.Q:.1f}")
        print(f"   Bandwidth: {self.system.bandwidth:.3f} Hz")
        print(f"   Stored energy at 5A: {self.system.E_stored:.3f} J")
        
        # Calculate null current
        I_null = self.find_null_current()
        
        print(f"\n🎯 Null Field Generation:")
        if abs(I_null) <= self.system.I_max:
            print(f"   Achievable with I = {I_null:.3f} A")
            print(f"   At resonance, field enhanced by Q factor")
            print(f"   Effective nulling current: {I_null/self.system.Q:.6f} A equivalent")
        else:
            print(f"   Need {abs(I_null):.1f}A for direct cancellation")
            print(f"   At resonance, Q={self.system.Q} provides enhancement")
            print(f"   Effective drive: {self.system.I_max}A × Q = {self.system.I_max * self.system.Q:.1f} A-equivalent")
        
        print("\n" + "="*60)

# Create integrated system
print("\nCreating integrated system...")
integrated = IntegratedSystem(params, meta_coil, system, mag_results)
integrated.summary()

# Calculate and plot null profile
null_pos, null_value = integrated.null_field_profile()
```
